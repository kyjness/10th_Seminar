```python
import pandas as pd

# 1. 데이터 로드 및 기본 정보 확인
# CSV 파일 불러오기
df = pd.read_csv('/content/drive/MyDrive/파이썬 머신러닝 완벽가이드/데이터/titanic/train.csv')
df.info()
```

```python
# 피처 추출 (독립변수 확인해보기)
df.columns
```

```python
# 데이터프레임의 상위 5개 행을 출력하여 데이터의 기본 구조를 확인하시오.
# (빈칸: df.______())
df.head(5)
```

```python
# 2. 피처 엔지니어링 1: 가족 수 변수 생성
# SibSp (형제자매 / 배우자 수) + Parch (부모 / 자녀 수) 값을 더하여 새로운 가족 수 변수를 만드시오.
# 변수 이름은 'Family_Size'로 지정하세요.
df['Family_Size'] = df['SibSp'] + df['Parch']
# Family_Size 변수의 분포를 확인하고, 전체적인 분포가 어떤지 설명하시오.
df['Family_Size'].value_counts()
```

```python
# 3. 피처 엔지니어링 2: 이름에서 호칭 추출
# 이름(`Name`)에서 Mr, Mrs, Miss 등 호칭을 추출하여 새로운 'Title' 변수를 생성하세요.
# 힌트: 이름 문자열에서 콤마(,)와 점(.) 사이의 텍스트를 추출하면 됨.
# (빈칸: df['Title'] = df['Name'].str.extract('_____', expand=False))
df['Title'] = df['Name'].str.extract(r',\s*([^\.]+)\.', expand=False)
# 생성된 Title 변수의 상위 5개 값을 출력하고, 다양한 호칭이 어떻게 구성되어 있는지 확인하세요.
print(df['Title'].head(5))
print(df['Title'].value_counts())
```

```python
# 4. 피처 엔지니어링 3: 나이 그룹화
# Age 변수의 결측값을 중간값으로 채우고, 나이 범주를 나눠서 'Age_Group' 변수를 생성하세요.
# 예: 0-12 (Child), 13-18 (Teenager), 19-35 (Young_Adult), 36-60 (Adult), 60 이상 (Senior)
# (빈칸: df['Age'] = df['Age'].fillna(df['______.median()']))
df['Age'] = df["Age"].fillna(df['Age'].median())
# 나이를 그룹화하는 함수 작성
def age_group(age):  # age_group 함수를 정의합니다.
    if age <= 12:
        return 'Child'
    elif age <= 18:
        return 'Teenager'
    elif age <= 35:
        return 'Young_Adult'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

# age_group 함수를 사용하여 새로운 'Age_Group' 변수 생성
# (빈칸: df['Age_Group'] = df['Age'].apply(____))
df['Age_Group'] = df['Age'].apply(age_group)
# 생성된 Age_Group 변수의 분포를 확인하세요.
# (빈칸: print(df['______'].value_counts()))
print(df['Age_Group'].value_counts())
```

```python
# 5. 피처 엔지니어링 4: 요금 범주화
# Fare 변수를 기준으로 값이 높은 승객과 낮은 승객을 구분하기 위해 4개의 구간으로 나누어 'Fare_Bin' 변수를 생성하세요.
# 힌트: pd.qcut() 함수를 사용하여 구간 나누기
# (빈칸: df['Fare_Bin'] = pd.qcut(df['____'], _____))
df['Fare_Bin'] = pd.qcut(df['Fare'], 4)
# Fare_Bin 변수의 빈도를 출력하고, 값의 분포를 확인하세요.
# (빈칸: print(df['______'].value_counts()))
df['Fare_Bin'].value_counts()
```

```python
# 6. 피처 엔지니어링 5: 객실 여부
# Cabin 변수가 결측인 경우와 아닌 경우를 구분하여 'Cabin_Known' 변수를 생성하세요.
# Cabin 값이 존재하면 1, 결측이면 0으로 설정하세요.
# (빈칸: df['Cabin_Known'] = df['Cabin'].notna().astype(____))
df['Cabin_Known'] = df['Cabin'].notna().astype(int)

# 생성된 Cabin_Known 변수의 빈도를 출력하고, 객실 여부와 생존율 간의 관계를 분석하세요.
# (빈칸: print(df['______'].value_counts()))
print(df['Cabin_Known'].value_counts())
print(df.groupby('Cabin_Known')['Survived'].mean())
```

```python
# 7. 피처 엔지니어링 6: 승선 지역
# 'Embarked' 변수를 더미 변수로 변환하여 승선 지역(S, C, Q)을 기준으로 각각의 더미 변수를 생성하세요.
# (빈칸: df = pd.get_dummies(df, columns=['____'], drop_first=True))
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
# 변환된 데이터프레임의 상위 5개 행을 출력하여 더미 변수 생성 결과를 확인하세요.
# (빈칸: df.______())
df.head(5)
```

5)데이터 분포 정규화:

### 2. 새로 생성한 변수에 대해 간단히 설명하고 모델의 성능이 향상된다면 그 이유가 무엇일지 설명하시오.

<새로 생성한 변수>

1.Family_Size:

SibSp(형제자매/배우자 수)와 Parch(부모/자녀 수)를 합한 변수입니다. 이 변수는 승객이 얼마나 많은 가족과 함께 탑승했는지를 나타냅니다. 가족 수가 많을수록 생존 가능성에 영향을 미칠 수 있으며, 이 변수는 모델에 중요한 피처로 작용할 수 있습니다.

2.Title:

이름에서 Mr., Mrs., Miss. 등과 같은 호칭을 추출한 변수입니다. 호칭은 승객의 성별, 결혼 여부, 사회적 지위를 반영할 수 있어 생존율 예측에 도움이 될 수 있습니다. 예를 들어, "Mr"는 남성 승객임을 나타내며, "Mrs"는 결혼한 여성임을 나타낼 수 있습니다.

3.Age_Group:

Age 변수를 기준으로 나이를 그룹화한 변수입니다. Child, Teenager, Young_Adult, Adult, Senior로 나누어 승객의 나이에 따른 생존율을 예측할 수 있습니다. 어린이와 노약자가 더 높은 생존율을 가질 가능성이 있습니다.

4.Fare_Bin:

Fare 변수를 4개의 구간으로 나누어 승객의 지불 요금을 범주화한 변수입니다. 요금이 높은 승객일수록 상류층일 가능성이 높고, 더 안전한 객실에 배치되었을 가능성이 있어 생존율에 영향을 미칠 수 있습니다.

5.Cabin_Known:

Cabin 변수가 결측인지 여부를 나타내는 변수입니다. 객실 정보가 있는 승객은 상위 클래스에 있었을 가능성이 있고, 이는 생존율과 관련이 있을 수 있습니다.

6.Embarked 더미 변수:

승선 지역을 나타내는 Embarked 변수를 더미 변수로 변환한 것입니다. 승선 지역에 따라 생존율에 차이가 있을 수 있습니다. 예를 들어, 특정 지역에서 탑승한 승객들이 더 안전한 구역에 배치되었을 가능성이 있습니다.

<모델 성능 향상 이유>

1.유의미한 정보 추가:

새로 생성된 변수들은 모델이 학습할 수 있는 추가적인 정보를 제공합니다. 예를 들어, Family_Size는 탑승한 가족의 수가 생존에 영향을 미칠 수 있고, Title 변수는 승객의 성별 및 사회적 지위와 관련이 있어 생존율 예측에 도움이 됩니다. 이러한 변수를 통해 모델이 더 많은 특징을 학습할 수 있게 되어 예측 성능이 향상될 수 있습니다.

2.데이터의 해석력 증가:

나이, 요금, 가족 수와 같은 변수들을 그룹화하거나 범주화함으로써 모델이 데이터에서 패턴을 더 잘 파악할 수 있게 됩니다. 예를 들어, Age_Group과 같은 변수를 통해 나이에 따른 생존율 차이를 쉽게 파악할 수 있습니다.

3.범주형 변수의 처리:

범주형 변수를 더미 변수로 변환하면, 모델이 이를 수치적으로 처리할 수 있게 되어 성능이 향상됩니다. Embarked 변수는 승선 지역에 따른 생존율 차이를 반영할 수 있습니다.

4.결측치 처리:

Cabin_Known과 같이 결측 여부를 새로운 변수로 처리하면, 결측값이 가지는 의미를 추가적으로 모델에 반영할 수 있습니다. 이는 모델이 결측값에 따른 패턴을 더 잘 이해하도록 도울 수 있습니다.

5.과적합 방지:

데이터를 그룹화하거나 범주화하면 모델의 복잡성을 줄이고, 중요한 패턴을 더 잘 학습하게 할 수 있습니다. 이로 인해 과적합을 방지하고 일반화 성능을 높일 수 있습니다.
